```python
# Then, import data from Foursquare API

venues = pd.read_csv("coffee&tea_ottawa_venues.csv", encoding = 'ISO-8859-1')

print(venues.shape)
venues.head()
```

```python
# one hot encoding
venues_onehot = pd.get_dummies(venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
venues_onehot['alias'] = venues['Store']

# move neighborhood column to the first column
fixed_columns = [venues_onehot.columns[-1]] + list(venues_onehot.columns[:-1])
venues_onehot = venues_onehot[fixed_columns]

venues_onehot.head()
```

```python
# Let's print each neighborhood along with the top 10 most common venues

venues_grouped = venues_onehot.groupby('alias').mean().reset_index()

num_top_venues = 10

for hood in venues_grouped['alias']:
    print("----"+hood+"----")
    temp = venues_grouped[venues_grouped['alias'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# write a function to sort the venues in descending order

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['alias']

for ind in np.arange(num_top_venues):
    
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
        
    except:
        
        columns.append('{}th Most Common Venue'.format(ind+1))
        

# create a new dataframe
venues_sorted = pd.DataFrame(columns=columns)
venues_sorted['alias'] = venues_grouped['alias']

for ind in np.arange(venues_grouped.shape[0]):
    
    venues_sorted.iloc[ind, 1:] = return_most_common_venues(venues_grouped.iloc[ind, :], num_top_venues)

venues_sorted.to_csv("neighbors.csv", index = False)
venues_sorted.head()
```

```python
# prepare for running a clustering

from sklearn.cluster import KMeans

# set number of clusters
kclusters = 5
clustering = venues_grouped.drop('alias', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add clustering labels
venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

venues_merged = df_store

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
venues_merged = venues_merged.join(venues_sorted.set_index('alias'), on='alias')

venues_merged.head() # check the last columns!
```

```python
# pick up the variables

df_regression = df_store[['alias', 'tag', 'if_popular', 'if_ranked', 'latitude', 'longitude']]

from math import radians, cos, sin, asin, sqrt 

def calcDistance(Lat_A, Lng_A, Lat_B, Lng_B):
    EARTH_RADIUS = 6378.137;
    radLat1 = radians(Lat_A)
    radLat2 = radians(Lat_B)
    a=radLat1-radLat2
    b=radians(Lng_A)-radians(Lng_B)
 
    s = 2 * asin(sqrt(pow(sin(a/2),2)+cos(radLat1)*cos(radLat2)*pow(sin(b/2),2)));  
    s = s * EARTH_RADIUS;  
    #s = Math.round(s * 10000d) / 10000d;  
    return s;
```

```python
df_regression['if_close_university'] = 0

for j in range(len(university)):

    for i in range(len(df_regression)):
    
        df_regression.loc[i, 'latitude']
        df_regression.loc[i, 'longitude']
        university.loc[j, 'latitude']
        university.loc[j, 'longitude']
        
        # km
        distance = calcDistance(df_regression.loc[i, 'latitude'], df_regression.loc[i, 'longitude'], 
                     university.loc[j, 'latitude'], university.loc[j, 'longitude'])
        
        if df_regression.loc[i, 'if_close_university'] == 0:
            
            if distance <= 0.5:
                
                df_regression.loc[i, 'if_close_university'] = 1
```

```python
df_regression['if_close_shopping'] = 0

for j in range(len(shopping)):

    for i in range(len(df_regression)):
    
        df_regression.loc[i, 'latitude']
        df_regression.loc[i, 'longitude']
        shopping.loc[j, 'latitude']
        shopping.loc[j, 'longitude']
        
        # km
        distance = calcDistance(df_regression.loc[i, 'latitude'], df_regression.loc[i, 'longitude'], 
                     shopping.loc[j, 'latitude'], shopping.loc[j, 'longitude'])
        
        if df_regression.loc[i, 'if_close_shopping'] == 0:
            
            if distance <= 0.5:
                
                df_regression.loc[i, 'if_close_shopping'] = 1        
```

```python
for i in range(len(df_regression)):
    
    if any(venues_sorted['alias'] == df_regression.loc[i, 'alias']):
        
        temp = venues_sorted[venues_sorted['alias'] == df_regression.loc[i, 'alias']]
        temp2 = temp.index[0]

        df_regression.loc[i, 'if_competitor'] = venues_sorted.loc[temp2, 'if_competitor']
        df_regression.loc[i, 'Cluster Labels'] = int(venues_sorted.loc[temp2, 'Cluster Labels'])

df_regression.dropna(subset=['if_competitor'], inplace=True)
df_regression = df_regression.reset_index(drop = True)
```

```python
import scipy

lst = ['tag', 'if_close_university', 'if_close_shopping', 'if_competitor', 'Cluster Labels']
lst2 = ['if_popular','if_ranked']
kendall_pvalue = []
kendall_corr = []
name_lst = []

for y in lst2:
    
    for item in lst:
        
        variable_name = item + "_" + y
        pvalue = scipy.stats.kendalltau(df_regression[item], df_regression[y]).pvalue
        corr = scipy.stats.kendalltau(df_regression[item], df_regression[y]).correlation
        kendall_pvalue.append(pvalue)
        kendall_corr.append(corr)
        name_lst.append(variable_name)

temp = {
        'variable_pair': name_lst,
        'correlation': kendall_corr,
        'p-value': kendall_pvalue
        
        }
corr_table = pd.DataFrame(temp)
corr_table = corr_table.sort_values('p-value', ascending=True).reset_index(drop = True) 
```
